```python
import pandas as pd
import json
import urllib3
import ast
from bs4 import BeautifulSoup
import codecs
import unicodedata
import binascii 
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
dict_columns = ['crew']

def text_to_dict(df):
    for column in dict_columns:
        df[column] = df[column].apply(lambda x: {} if pd.isna(x) else ast.literal_eval(x) )
    return df
train = text_to_dict(train)

test = text_to_dict(test)
```

```python
def get_director(j):
    for elem in j:
        if elem['job'] == 'Director':
            return elem['name']
    return ''

def get_score(name, f, id):
    try:
        name = name.lower().replace(' ', '_')
        name = unicodedata.normalize('NFKD', name).encode('ASCII', 'ignore').decode('ASCII')
        url = 'https://www.rottentomatoes.com/celebrity/' + name
        print(url + '|' + str(id))
        http = urllib3.PoolManager()
        r = http.request('GET', url)
        soup = BeautifulSoup(r.data, 'html.parser')

        scores = (soup.find_all('span',{'class':'tMeterScore'}))

        total_score = 0 
        for s in scores:
            total_score += int(s.text.split('%')[0])

        f.write(str(id) + ',' + str(total_score/len(scores)) + '\n')
        f.flush()
        return total_score/len(scores)

    except:
        f.write(str(id) + ',' + 'PLACE' + '\n')
        f.flush()
        print("error")
    
f = open('director', 'a')
for i in range(3000, 3000):
    get_score(get_director(train['crew'][i]), f, i + 1)
#train['director_score'] = train['crew'].apply(lambda x : get_score(get_director(x)))
```

```python
f = open('director_test', 'a')
for i in range(0, 4404):
    get_score(get_director(test['crew'][i]), f, i + 1)
```
